# Make script to automate adding the positions

```python
import pandas as pd
import json
```

```python
# Revised script for better efficiency, each run will work for one year, so that the output can be more easily managed
# poor readability cos edge case logic, but much better functioning than before
years = [2021, 2022, 2023, 2024]

for year in years:
    filtered_list = [] # -> List of dictionaries to store necessary values of DF 

    # Collect all games and their participants within a season
    with open(f"wnba_betting_data/{year}/player_game_stats_{year}.csv") as f:
        original_df = pd.read_csv(f)
        for index, row in original_df.iterrows():
            temp_dict = {"game_id": row["game_id"], "player_id": row["player_id"], "home_away": row["home_away"]}
            filtered_list.append(temp_dict)

    game_id = None if not filtered_list else filtered_list[0]["game_id"]
    game_batch = []
    final_form = []
        
    for i in range(len(filtered_list)):
        
        if i == (len(filtered_list) - 1):
            game_batch.append(filtered_list[i])
            game_id = "Activate Processing" # To handle edge case for last game (1)

        if game_id != filtered_list[i]["game_id"]:
            if i == (len(filtered_list) - 1): # To handle edge case for last game (2)
                game_id = filtered_list[i]['game_id'] 

            with open(f"wnba_betting_data/{year}/game_{game_id}.json") as f:
                data = json.load(f)
                temp_df_home = pd.json_normalize(data, record_path=["home", "players"])
                temp_df_away = pd.json_normalize(data, record_path=["away", "players"])
                for row in game_batch:
                    player_id = row["player_id"]
                    if row["home_away"] == "home":
                        position_element = temp_df_home.query(f"id == '{player_id}'")["position"].iloc[0]
                        temp_dict = {"game_id": game_id, "player_id": player_id, "home_away": row["home_away"], "position": position_element}
                        final_form.append(temp_dict)
                    elif row["home_away"] == "away":
                        position_element = temp_df_away.query(f"id == '{player_id}'")["position"].iloc[0]
                        temp_dict = {"game_id": game_id, "player_id": player_id, "home_away": row["home_away"], "position": position_element}
                        final_form.append(temp_dict)
                    
            game_id = filtered_list[i]["game_id"]
            game_batch.clear()
            game_batch.append(filtered_list[i]) # append this row or else it will be skipped

        else:
            game_batch.append(filtered_list[i]) 

    final_form = pd.DataFrame(final_form)

    player_game_stats_positions = pd.concat([original_df, final_form.filter(items=["position"])], axis=1)
    player_game_stats_positions.to_csv(f"wnba_betting_data/{year}/player_game_stats_positions_{year}.csv", index=False)

```
